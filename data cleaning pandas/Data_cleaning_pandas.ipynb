```python
import pandas as pd
```

```python
dataset= pd.read_excel('Customer Call List.xlsx')
dataset.head()
```

Checking for duplicate values in each rows

```python
dataset.duplicated()
```

```python
# droping the duplicates
dataset=dataset.drop_duplicates()
```

```python
dataset.duplicated().value_counts() # taking a count of duplicate values.Zero here, since dataset is dup-dropped.
```

```python
dataset['Last_Name']=dataset['Last_Name'].str.lstrip("...")  #used to remove non needed characters on last name(left-side).
```

```python
dataset
```

```python
dataset['Last_Name']=dataset['Last_Name'].str.lstrip("/")  #used to remove '/' character on last name(left-side).
dataset
```

```python
dataset['Last_Name']=dataset['Last_Name'].str.rstrip("_") 
dataset
```

Working with phone no.

```python
dataset['Phone_Number']=dataset['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
dataset.head(10)
```

```python
 dataset.dtypes  #checking the datatype of various column values.
```

Converting the phone_number col data to string type and apply lambda function to add spaces in each no.

```python
dataset["Phone_Number"]=dataset["Phone_Number"].apply(lambda x: str(x))
dataset['Phone_Number']=dataset['Phone_Number'].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10])
dataset.head(10)
```

```python
dataset['Phone_Number']=dataset['Phone_Number'].str.replace('nan--','')
dataset.head(10)
```

Arranging the address in three seperate columns

```python
dataset['Address'].str.split(',',2,expand=True).head(10)
```

```python
dataset[["Street","State","Pincode"]]=dataset['Address'].str.split(',',2,expand=True)
dataset.head(5)
```

```python
#dataset.columns
```

```python
#Dropping the existing address column 

dataset.drop(columns="Address",inplace=True,axis=1)
dataset
```

```python
dataset["Do_Not_Contact"]=dataset["Do_Not_Contact"].str.replace('Yes','Y')
dataset["Do_Not_Contact"]=dataset["Do_Not_Contact"].str.replace('No','N')
# Having a single character to represent yes or no for col. Do_Not_contact and .
```

```python
dataset["Paying Customer"]=dataset["Paying Customer"].str.replace('Yes','Y')
dataset["Paying Customer"]=dataset["Paying Customer"].str.replace('No','N')
# Having a single character to represent yes or no for col. Do_Not_contact and .
```

```python
dataset.fillna('') #substituting Nan Values by empty space.
```

```python
dataset
```

```python
dataset=dataset.replace('N/a','')
dataset=dataset.replace('Na--','')
dataset['Do_Not_Contact']=dataset['Do_Not_Contact'].replace('','N')
dataset.head(7)
```

```python
for x in dataset.index:  # droping the rows that have a do_not_cnct as True.
    if dataset.loc[x,'Do_Not_Contact']=='Y':
        dataset.drop(x,inplace=True)
        
dataset  #check the index to verify. rows with index:1,3 and 18 are dropped.  
```

```python
for x in dataset.index:  # droping the rows that have a do_not_cnct as True.
    if dataset.loc[x,'Phone_Number']=='':
        dataset.drop(x,inplace=True)
        
dataset  
```
